### Работа с NaN
Часто данные содержат пропущенные значения, которые представляются как NaN (Not a Number). Pandas предоставляет удобные методы для работы с ними.

Проверка на наличие NaN. Методы fillna и dropna возвращают новые Dataframe, проверьте нет ли в них NaN

```python
# Проверка на наличие NaN в DataFrame
# Используйте метод .isna()
titanic_df.isna().sum()
```

Заполнение NaN

```python
# Заполнение NaN определенным значением (например, нулем)
# Используйте метод .fillna()
titanic_filled = titanic_df.fillna(0)

# Проверим, что в новом DataFrame нет NaN
titanic_filled.isna().sum()
```

Удаление строк с NaN

```python
# Удаление строк, содержащих NaN
# Используйте метод .dropna()
titanic_no_na = titanic_df.dropna()

# Проверим, что в этом DataFrame тоже нет NaN
titanic_no_na.isna().sum()
```

Проверьте, есть ли пропущенные значения в данных и заполните их нулями.
Выведите первые 10 строк данных.
Выберите только те строки, где значение в столбце 'Age' больше 30.
Отсортируйте данные по столбцу 'Fare' в порядке убывания.
Сгруппируйте данные по столбцу 'Pclass' и вычислите средний возраст ('Age') для каждого класса.

```python
# 1. Прочитайте данные из файла 'titanic.csv'
titanic_df = pd.read_csv('titanic.csv')

# 2. Проверьте, есть ли пропущенные значения в данных
print(titanic_df.isna().sum())

# 3. Заполните пропущенные значения нулями
titanic_df = titanic_df.fillna(0)

# 4. Выведите первые 10 строк данных
print(titanic_df.head(10))

# 5. Выберите только те строки, где Age > 30
age_over_30 = titanic_df[titanic_df['Age'] > 30]
print(age_over_30.head())

# 6. Отсортируйте данные по Fare в порядке убывания
sorted_by_fare_desc = titanic_df.sort_values(by='Fare', ascending=False)
print(sorted_by_fare_desc.head())

# 7. Сгруппируйте по Pclass и вычислите средний Age
mean_age_by_pclass = titanic_df.groupby('Pclass')['Age'].mean()
print(mean_age_by_pclass)
```
