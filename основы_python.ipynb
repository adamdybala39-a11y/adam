#Задача 1
Написать функцию на вход которой подается строка, состоящая из латинских букв.
Функция должна вернуть количество гласных букв (a, e, i, o, u) в этой строке. Написать тесты для кода

```python
def count_vowels(s):
    vowels = 'aeiouAEIOU'
    count = 0
    for char in s:
        if char in vowels:
            count += 1
    return count

# Тесты
def test_count_vowels():
    # Тест 1: строка с гласными
    assert count_vowels("hello") == 2, "Test 1 failed"
    
    # Тест 2: строка без гласных
    assert count_vowels("xyz") == 0, "Test 2 failed"
    
    # Тест 3: пустая строка
    assert count_vowels("") == 0, "Test 3 failed"
    
    # Тест 4: все гласные
    assert count_vowels("aeiou") == 5, "Test 4 failed"
    
    # Тест 5: с заглавными буквами
    assert count_vowels("HELLO") == 2, "Test 5 failed"
    
    print("All tests passed!")

test_count_vowels()
```

# Задача 2
Написать функцию на вход, которой подается строка. Функция должна вернуть true, если
каждый символ в строке встречается только 1 раз, иначе должна вернуть false. Написать тесты для кода

```python
def is_unique(s):
    return len(s) == len(set(s))

# Тесты
def test_is_unique():
    # Тест 1: все уникальные символы
    assert is_unique("abcdef") == True, "Test 1 failed"
    
    # Тест 2: повторяющиеся символы
    assert is_unique("hello") == False, "Test 2 failed"
    
    # Тест 3: пустая строка
    assert is_unique("") == True, "Test 3 failed"
    
    # Тест 4: один символ
    assert is_unique("a") == True, "Test 4 failed"
    
    # Тест 5: два одинаковых символа
    assert is_unique("aa") == False, "Test 5 failed"
    
    print("All tests passed!")

test_is_unique()
```

# Задача 3
Написать функцию, которая принимает положительное число и возвращает количество
бит равных 1 в этом числе. Написать тесты для кода

```python
def count_bits(n):
    count = 0
    while n:
        count += n & 1
        n >>= 1
    return count

# Альтернативный способ
def count_bits_alt(n):
    return bin(n).count('1')

# Тесты
def test_count_bits():
    # Тест 1: число 5 (бинарно 101)
    assert count_bits(5) == 2, "Test 1 failed"
    
    # Тест 2: число 1
    assert count_bits(1) == 1, "Test 1 failed"
    
    # Тест 3: число 7 (бинарно 111)
    assert count_bits(7) == 3, "Test 3 failed"
    
    # Тест 4: число 8 (бинарно 1000)
    assert count_bits(8) == 1, "Test 4 failed"
    
    # Тест 5: число 15 (бинарно 1111)
    assert count_bits(15) == 4, "Test 5 failed"
    
    print("All tests passed!")

test_count_bits()
```

# Задача 4
Написать функцию, которая принимает положительное число. Функция должна вернуть то,
сколько раз необходимо перемножать цифры числа или результат перемножения, чтобы
получилось число состоящее из одной цифры.
Например, для входного числа:
· 39 функция должна вернуть 3, так как 3*9=27 => 2*7=14 => 1*4=4
· 4 функция должна вернуть 0, так как число уже состоит из одной цифры
· 999 функция должна вернуть 4, так как 9*9*9=729 => 7*2*9=126 => 1*2*6=12 =>
1*2=2. Написать тесты для кода

```python
def magic(n):
    steps = 0
    while n >= 10:
        product = 1
        while n > 0:
            product *= n % 10
            n //= 10
        n = product
        steps += 1
    return steps

# Тесты
def test_magic():
    # Тест 1: 39
    assert magic(39) == 3, "Test 1 failed (3*9=27 => 2*7=14 => 1*4=4)"
    
    # Тест 2: 4
    assert magic(4) == 0, "Test 2 failed (одна цифра)"
    
    # Тест 3: 999
    assert magic(999) == 4, "Test 3 failed (9*9*9=729 => 7*2*9=126 => 1*2*6=12 => 1*2=2)"
    
    # Тест 4: 10
    assert magic(10) == 1, "Test 4 failed (1*0=0)"
    
    # Тест 5: 25
    assert magic(25) == 2, "Test 5 failed (2*5=10 => 1*0=0)"
    
    print("All tests passed!")

test_magic()
```

# Задача 5
Написать функцию, которая принимает два целочисленных вектора одинаковой длины и
возвращает среднеквадратическое отклонение двух векторов. Написать тесты для кода

```python
def mse(pred, true):
    if len(pred) != len(true):
        raise ValueError("Vectors must have the same length")
    
    sum_squared_errors = sum((p - t) ** 2 for p, t in zip(pred, true))
    return sum_squared_errors / len(pred)

# Тесты
def test_mse():
    # Тест 1: идеальное совпадение
    assert mse([1, 2, 3], [1, 2, 3]) == 0, "Test 1 failed"
    
    # Тест 2: простой случай
    result = mse([1, 2, 3], [2, 3, 4])
    expected = (1 + 1 + 1) / 3
    assert result == expected, "Test 2 failed"
    
    # Тест 3: отрицательные числа
    result = mse([1, 2], [-1, 0])
    expected = (4 + 4) / 2
    assert result == expected, "Test 3 failed"
    
    print("All tests passed!")

test_mse()
```

# Задача 6
Написать функцию, принимающая целое положительное число. Функция должна вернуть
строку вида "(n1**p1)(n2**p2)…(nk**pk)" представляющая разложение числа на простые
множители (если pi == 1, то выводить только ni).
Например, для числа 86240 функция должна вернуть "(2**5)(5)(7**2)(11)". Написать тесты для кода

```python
def prime_factorization(n):
    factors = {}
    d = 2
    
    while d * d <= n:
        while n % d == 0:
            factors[d] = factors.get(d, 0) + 1
            n //= d
        d += 1
    
    if n > 1:
        factors[n] = factors.get(n, 0) + 1
    
    result = ""
    for prime in sorted(factors.keys()):
        if factors[prime] == 1:
            result += f"({prime})"
        else:
            result += f"({prime}**{factors[prime]})"
    
    return result

# Тесты
def test_prime_factorization():
    # Тест 1: 86240
    assert prime_factorization(86240) == "(2**5)(5)(7**2)(11)", "Test 1 failed"
    
    # Тест 2: простое число
    assert prime_factorization(7) == "(7)", "Test 2 failed"
    
    # Тест 3: 12
    assert prime_factorization(12) == "(2**2)(3)", "Test 3 failed"
    
    # Тест 4: 2
    assert prime_factorization(2) == "(2)", "Test 4 failed"
    
    print("All tests passed!")

test_prime_factorization()
```

# Задача 7
Написать функцию, принимающая целое число n, задающее количество кубиков. Функция
должна определить, можно ли из данного кол-ва кубиков построить пирамиду, то есть
можно ли представить число n как 1^2+2^2+3^2+…+k^2. Если можно, то функция должна
вернуть k, иначе строку "It is impossible". Написать тесты для кода

```python
def pyramid(number):
    k = 1
    sum_of_squares = 0
    
    while sum_of_squares < number:
        sum_of_squares += k * k
        if sum_of_squares == number:
            return k
        k += 1
    
    return "It is impossible"

# Тесты
def test_pyramid():
    # Тест 1: 1 = 1^2
    assert pyramid(1) == 1, "Test 1 failed"
    
    # Тест 2: 5 = 1^2 + 2^2
    assert pyramid(5) == 2, "Test 2 failed"
    
    # Тест 3: 14 = 1^2 + 2^2 + 3^2
    assert pyramid(14) == 3, "Test 3 failed"
    
    # Тест 4: невозможно
    assert pyramid(10) == "It is impossible", "Test 4 failed"
    
    # Тест 5: 55 = 1^2 + 2^2 + 3^2 + 4^2 + 5^2
    assert pyramid(55) == 5, "Test 5 failed"
    
    print("All tests passed!")

test_pyramid()
```

# Задача 8
Функция принимает на вход положительное число и определяет является ли оно сбалансированным, т.е. сумма цифр до средних равна сумме цифр после. Средними в случае нечетного числа цифр считать одну цифру, в случае четного - две средних. Написать тесты для кода

```python
def is_balanced(n):
    s = str(n)
    length = len(s)
    
    if length % 2 == 0:
        # Четное число цифр
        mid = length // 2
        left_sum = sum(int(d) for d in s[:mid])
        right_sum = sum(int(d) for d in s[mid:])
    else:
        # Нечетное число цифр
        mid = length // 2
        left_sum = sum(int(d) for d in s[:mid])
        right_sum = sum(int(d) for d in s[mid+1:])
    
    return left_sum == right_sum

# Тесты
def test_is_balanced():
    # Тест 1: 121 (четное число цифр: 1=1)
    assert is_balanced(121) == True, "Test 1 failed"
    
    # Тест 2: 1221 (четное число цифр: 1+2=2+1)
    assert is_balanced(1221) == True, "Test 2 failed"
    
    # Тест 3: 123 (нечетное число цифр: 1!=3)
    assert is_balanced(123) == False, "Test 3 failed"
    
    # Тест 4: 1000 (четное число цифр: 1+0=0+0)
    assert is_balanced(1000) == False, "Test 4 failed"
    
    # Тест 5: 11 (четное число цифр: 1=1)
    assert is_balanced(11) == True, "Test 5 failed"
    
    print("All tests passed!")

test_is_balanced()
```
